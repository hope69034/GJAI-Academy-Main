```python
import random   #랜덤가방임폴트
answer = random.randrange(1,101)  #정답값을 랜덤으로 줌
list_1 = []   # 인풋순서를 출력하기 위한 빈리스트 선언
while True:        # 에러가 발생해도 정답을 맞출 때까지 멈추지 않기 위한 와일트루문
     try:          #에러 처리를 위한 트라이익셉문            
            inp=input('1이상 100이하의 정수를 입력하세요: ')    # 인풋 받기  
            if int(inp) != answer :        # 인풋이 정답이 아니라면~                            
               if   1 <= int(inp) <= 100 :       #인풋이 1~100 범위일 때
                  if answer > int(inp) :         # 인풋이 정답보다 작으면 업을 출력하고 인풋순서리스트에 넣음
                        print("Up")
                        list_1.append(inp)
                  else:             
                        print("Down")            # 인풋이 정답보다 크면 다운을 출력하고 인풋 순서리스트에 넣음
                        list_1.append(inp)  
               else:
                    list_1.append(inp)       # 인풋이 1~100사이의 정수가 아닌 정수라면 안내문구 출력 후 재입력
                    print('지정범위를 벗어난 정수입니다. 1이상 100이하의 정수를 입력하세요')
            else:               # 인풋이 정답이라면
               list_1.append(inp)    # 인풋순서리스트에 넣고
               x = len(list_1)         # 인풋순서리스트의 길이를 x 로 선언 == 인풋 넣은 횟수
               print(f"{x}회째 정답!")    # x 회째 정답 출력
               print('*인풋 순서*')      #인풋순서 안내문구 출력
               for xx in range(x):       #인풋순서를 하나씩 출력하기 위해 인풋순서리스트의 길이만큼 반복하는 반복문
                  if  xx+1 == x :     # xx에 0부터 들어가기 때문에 +1을 해주고 ==x로 현재들어온 xx가 마지막 값인지를 판단
                                      # x는 인풋넣은수, 리스트의길이기 때문에                          
                      print(xx+1,'번째 인풋은', list_1[xx],'정답입니다')   # 리스트[인덱스값]으로 매인풋값을 불러오고
                                                                           #  마지막값은 정답이기 때문에 정답입니다를 출력
                  else:        # xx가 마지막 값이 아니라면 정답이 아니기 때문에 입니다만 출력
                      print(xx+1,'번째 인풋은', list_1[xx],'입니다') 
               break     #  정답을 맞혔기 때문에 브레이크로 코드 전부를 탈출하며 완전히 종료
     except:                # 인트가 아닌 것을 입력하여 에러발생시 처리를 위한 익셉문
                if inp[0].isspace():     # 인풋을 공백으로 줄 경우 공백을 입력했다는 안내를 하기 위한 이프문
                                         # str.isspace() 로 인풋의 0인덱 값에 공백이 있다면 ~ 조건생성으로
                                        # 공백이 하나부터 무한대라도 처리할 수 있도록 함
                                        # inp[0]으로 인풋의 0인덱스값이 공백이 있다면 True 없다면 False 를 리턴  
                   print('공백을 입력했습니다. 1이상 100이하의 정수를 입력하세요') #공백입력시 공백을입력했습니다라고 안내
                   list_1.append(inp)       #공백을 인풋순서리스트에 넣음
                else:     #(정수가 아니고) 공백이 아닌 인풋으로 에러 발생 시 ex)특수문자
                      print('정수가 아닙니다. 1이상 100이하의 정수를 입력하세요') # 정수가 아님을 안내
                      list_1.append(inp)    # 인풋순서리스트에 넣음
```

```python
####################       완성    

import random   #랜덤가방 임폴트
answer = random.randrange(1,101)  #정답값을 랜덤으로 줌
list_1 = []   # 인풋순서를 출력하기 위한 빈리스트 선언
while True:        # 에러가 발생해도 정답을 맞출 때까지 멈추지 않기 위한 와일트루문
     try:          #에러 처리를 위한 트라이익셉문            
            inp=input('1이상 100이하의 정수를 입력하세요: ')    # 인풋 받기  
            if int(inp) != answer :        # 인풋이 정답이 아니라면~                            
               if   1 <= int(inp) <= 100 :       #인풋이 1~100 범위일 때
                  if answer > int(inp) :         # 인풋이 정답보다 작으면  
                        print("Up")              # 업을 출력하고
                        list_1.append(inp)       # 인풋순서리스트에 넣음
                  else:             
                        print("Down")            # 인풋이 정답보다 크면 다운을 출력하고 
                        list_1.append(inp)       # 인풋 순서리스트에 넣음
               else:
                    list_1.append(inp)       # 인풋이 1~100사이의 정수가 아닌 정수라면 안내문구 출력 후 재입력
                    print('지정범위를 벗어난 정수입니다. 1이상 100이하의 정수를 입력하세요')
            else:               # 인풋이 정답이라면
               list_1.append(inp)    # 인풋순서리스트에 넣고
               x = len(list_1)         # 인풋순서리스트의 길이를 x 로 선언 == 인풋 넣은 횟수
               print(f"{x}회째 정답!")    # x 회째 정답 출력
               print('*인풋 순서*')      #인풋순서 안내문구 출력
               for xx in range(x):       #인풋순서를 하나씩 출력하기 위해 인풋순서리스트의 길이만큼 반복하는 반복문
                  if  xx+1 == x :     # xx에 0부터 들어가기 때문에 +1을 해주고 ==x로 현재들어온 xx가 마지막 값인지를 판단
                                      # x는 인풋넣은수, 리스트의길이기 때문에                          
                      print(xx+1,'번째 인풋은', list_1[xx],'정답입니다')   # 리스트[인덱스값]으로 매인풋값을 불러오고
                                                                           #  마지막값은 정답이기 때문에 정답입니다를 출력  
                  else:        # 마지막값이 아닌 경우 
                       if   list_1[xx].isspace() :    # 마지막값이 아니고 전부공백이면     
                          print(xx+1,'번째 인풋은', list_1[xx],'1개 이상의 공백을 입력했습니다') #공백을 인풋 했다는 것을 명시
                       else:   # 마지막값이 아니고 전부 공백은 아닌 경우
                          if list_1[xx] == '' :      #  인풋이 enter 인 경우 enter키를 누른 것을 명시
                                print(xx+1,'번째 인풋은', 'enter키를 입력하셨습니다')
                          else:                      #  인풋이 마지막값이 아니고 전부공백이 아니고 엔터가 아닌 경우
                                print(xx+1,'번째 인풋은', list_1[xx],'입니다')   
               break     #  정답을 맞혔기 때문에 브레이크로 코드 전부를 탈출하며 완전히 종료
     except:                # 인트가 아닌 것을 입력하여 에러발생시 처리를 위한 익셉문
                if inp.isspace():     # 인풋을 전부공백으로 줄 경우 공백을 입력했다는 안내를 하기 위한 이프문 
                                         # str.isspace() 로 인풋 전체의 값 전부가 공백이라면 ~ 조건생성으로, inp[인덱]삽입시 그 인덱만 검사
                                        # 공백이 하나부터 무한대라도 처리할 수 있도록 함 
                   print('전부 공백을 입력했습니다. 1이상 100이하의 정수를 입력하세요') #공백입력시 공백을입력했습니다라고 안내
                   list_1.append(inp)       #공백을 인풋순서리스트에 넣음
                else:     #(정수가 아니고) 공백이 아닌 인풋으로 에러 발생 시 ex)특수문자
                      if inp == '' :       # 인풋이 엔터 '' 키를 바로 눌렀는지 검사
                          print('입력하지 않고 엔터를 눌렀습니다. 1이상 100이하의 정수를 입력하세요')
                          list_1.append('')      # '' >  엔터입력을 인풋리스트로 추가
                      else:  
                          print('정수가 아닙니다. 1이상 100이하의 정수를 입력하세요') # 정수가 아님을 안내
                          list_1.append(inp)    # 인풋순서리스트에 넣음
```
