```python
#난수0~100 을 이용하여 백명의학생들의점수를만들고조건에맞게출력하는프로그램을만드시오


x=np.random.randint(0,101    ,100) #영에서백까지 랜덤으로  10개뽑는다      (시작,종료,몇개)

# 1. 평균이 50점이상이면 전원통과출력

print(x)
print()
print(np.mean(x))

if np.mean(x) >= 50:
    print('전원통과')

if np.all(x >= 50):               #전원50점이상일때
    print('학년1등반출력')

if np.any(x==100) :
    print('전교1등이있는반출력')  #한명이라도백점일때

```

```python
판다스  #데이터구조

판다스란
    파이썬의데이터분석라이브러리
        데이터테이블을다루는도구

    기본적으로 넘파이를사용
        판다스는넘파이를효율적으로사용하기위해인덱싱연산전처리등다양한함수제공

데이터프레임
    데이터테이블전체객체



^^^^^^^^^^^^^^


시리즈 (판다스의 시리즈)
    각열데이터를다루는객체

  피쳐의 시리즈    피쳐의 각 열데이터를 말함   1차원의 값

시리즈객체
    일반적으로하나의피쳐데이터를포함하는형태
    생성된데이터프레임안에포함될수잇음
    리스트딕셔너리엔디어레이등다양한데이터타입이시리즈객체형태로변환되기도함

시리즈객체를생성하면세가지요소생성
    데이터
        기존다른객체처럼값을저장하는요소
        시리즈데이터하나가 하나의자료형을가져야함 시리즈하나는 같은자료형으로 구성되어야 한다
        (데이터프레임[전체데이터]는 같은자료형이아니어도 된다는말)
    인덱스
        0부터시작하고숫자로만할당하는값
        시리즈객체에서는숫자문자열영외의값으로시작하는숫자
        순서가일정하지않은숫자를입력할수도있음
        시리즈객체에서는인덱스값의중복을허용
            인덱스이름을바꿀수잇다 첫인덱도 두번쨰인덱도 인덱스1로 정할수잇다
    데이터타입
        넘파이의데이터타입과일치
        판다스는넘파이의래퍼라이브러리
        넘파이의모든기능지원하고데이터타입도그대로적용

시리즈객체는넘파이배열의하위클래스
넘파이가지원하는어떠한데이터타입도지원
```
