```python
import matplotlib.pyplot as plt
X=range(10)
Y=range(10)
plt.plot(X,Y) #연결그래프함수
```

```python
파이플롯
  그래프를쌓아올올려서한번에                   그림 피겨 (바닥)-축 액시스(쌓아올림)
  (행,열,축위치)
```

```python
import numpy as np
c_Y=[np.cos(i) for i in X]
s_Y=[np.sin(i) for i in X]
plt.plot(s_Y) 
plt.plot(c_Y)
plt.show()   # 보여줘  (코랩이라안해도그냥출력)
```

```python
fig,ax=plt.subplots()
ax.plot(c_Y)
ax.set(title='cos_graph',xlabel='n' , ylabel='y')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)  #그림사이즈
fig.add_subplot(1,2,1)           #플랏 +엣스안붙임 하나 단수
fig.add_subplot(1,2,2) 
```

```python
fig,ax=plt.subplots(2,2)     # (가로로쪼개는수,세로로쪼개는수  1은 그대로)  #세로로 2개로 쪼갠다 가로는 x  


# x=np.linspace(-1,1,100)  #-1에서1까지 100개의원소만들기 동일한 간격으로 쪼개기
x=np.linspace(-1,1,100)
y_1=np.sin(x) #원소100개
y_2=np.cos(x)#원소100개
y_3=np.tan(x)#원소100개
y_4=np.exp(x)#원소100개


```

```python


#어디에 어떻게 넣을 수 잇을까
#왼쪽위- 영컴마영  오른쪽위 영컴마일...

fig,ax=plt.subplots(2,2)     # (가로로쪼개는수,세로로쪼개는수  1은 그대로)  #세로로 2개로 쪼갠다 가로는 x  


# x=np.linspace(-1,1,100)  #-1에서1까지 100개의원소만들기 동일한 간격으로 쪼개기
x=np.linspace(-1,1,100)
y_1=np.sin(x) #원소100개
y_2=np.cos(x)#원소100개
y_3=np.tan(x)#원소100개
y_4=np.exp(x)#원소100개
#ax 축
ax[0,0].plot(x,y_1)  #엑스컴마와이   엑스원소 백개 와이원소 백개   같아야함
ax[0,1].plot(x,y_2)  #저차원먼저바뀐다- 이름과 먼 곳
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
#코랩이라바로출력가능
plt.show()
```
