```python
# 데이터.인포()    함수
df.info()  #데이터의 각종 정보를 볼 수 있다
# Dtype 을 보면 정보가 맞지않다
# 오브젝 타입과 넘 타입을 아래서 바로잡아 준다
```

```python
# 타입을 바꾸는 함수  데이터.astype( )
oj_colums=['PassengerId','Pclass','Name','Sex','Ticket','Cabin','Embarked']
num_colums=['Age','SibSp','Parch','Fare']
for i in oj_colums:
  df[i]=df[i].astype(object)    #오브젝 시리즈를 오브젝타입으로 바꾼다
for i in num_colums:
  df[i]=df[i].astype(float)    #넘 시리즈를 실수 타입으로 바꾼다

df['Parch']=df['Parch'].astype(int)
df['SibSp']=df['SibSp'].astype(int) # 두 시리즈는 정수 타입으로 바꿨다
```

```python
df.info()  # 오브젝/정수/실수 타입으로 제대로 바뀐 것을 확인할 수 있다
```
