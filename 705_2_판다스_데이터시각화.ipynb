```python
판다스 병합

내부조인 - 교집합만 병합
완전조인 - 합집합

left_on 왼쪽조인   왼쪽테이블을 기준으로   왼+교집합
오른쪽조인                         오른+교집합

```

```python
d1={'c_id':['1','2','3','4','5'] ,'data':[10,20,30,40,50]}

df_l=DataFrame(d1)
df_l
```

```python
d2={'c_id':['3','4','5','6','7'] ,'data':[10,20,30,40,50],'data1':[100,200,300,400,500]}

df_r=DataFrame(d2)
df_r
```

```python
pd.merge(left=df_l,right=df_r,how='inner',on='c_id')   #판다스머지-병합 #이너 내부병합- 교집합
#c_id를 기준으로 교집합
```

```python
pd.merge(left=df_l,right=df_r,how='left',on='c_id') #왼쪽병합 #c_id기준
```

```python
pd.merge(left=df_l,right=df_r,how='right',on='c_id') 
```

```python
pd.merge(left=df_l,right=df_r,how='outer',on='c_id')  # 외부병합 합집함
```

```python
df_l.index=df_l.c_id
del df_l['c_id']        #인덱스를 시아이디로 만들기
df_l
```

```python
df_r.index=df_r.c_id
del df_r['c_id']
df_r
```

```python
pd.merge(left=df_l,right=df_r,how='outer',on='c_id') 
```
