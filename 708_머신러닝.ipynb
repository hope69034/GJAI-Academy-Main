```python
#연립방정식을 행렬로 풀기

# 2x-y =0      >>    2  -1    x =   0
#   x+y =3           1   1    y =   3

x=np.array([[2,-1],[1,1]])
xx=inv(x)
y=np.array([[0],[3]])
xx@y

```

```python
# 2x+y=1
# x-2y=4  의연립방정식을 행렬로 풀기

#2  3 x  1
#1 -2 y  4

x=np.array([[2,3],[1,-2]])    
y=np.array([[1],[4]])
inv(x)@y
```

```python

import pandas as pd
import numpy as np
#데이터로드 수집
data_d={
'이름':['홍길동','도우너',np.nan,'둘리','또치','희동이'], #결측치 넣기
'나이':[38,10,np.nan,3,5,0],
'성별':['남','남',np.nan,'남','여',np.nan],
'중간_점수':[50,np.nan,np.nan,3,5,np.nan],
'기말_점수':[25,np.nan,np.nan,60,50,10]
}

df =pd.DataFrame(data_d,columns=['이름','나이','성별','중간_점수','기말_점수'])
print(df)
df.groupby('성별')['기말_점수'].mean()
df.groupby('성별')['기말_점수'].transform('mean') 
#희동이성별이난이라 기말점수10이 nan으로 표시   위에서 성별기준 기말점수라
#원본수정은위코드로는안함
df.기말_점수.fillna(df.groupby('성별')['기말_점수'].transform('mean'))
```
