```python
# 맨아래 그래프 자르는 방법


import numpy as np
import matplotlib.pyplot as plt   # 그래프로 시각화하는 라이브러리
# 어레인지는 앞에 np가 필수
# np.arange ( a,b,c )     a부터 b까지 c스텝으로 뽑는다   # () 안에 숫자 하나만 있다면 0부터 그 숫자 미만까지 1스텝으로 뽑는다는 말이다
x=np.arange(-10,11,1)
f_x=x**2
# 피엘티플롯 x를 엑스축 f_x를 와이축으로 하는 선 그래프를 그린다
plt.plot(x,f_x)
# 피엘티쇼 (프린트같은역할, 코랩에선 없어도 자동 출력)
plt.show()

```

```python
from numpy.random.mtrand import random_integers
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt  # 시각화,그래프그리기
import numpy as np
import random

# 메이크_리그레션 : 가상의 선형회귀 분석용 데이터를 만드는 함수다  3개의 변수를 받고 
data, y, w = make_regression(n_samples=200, n_features=1, bias=100, noise=10,     coef=True,               random_state=1                )
                        #  표본 200개 ,     피처1차원     절편100   표준편차10   선형모형의계수도출력     함수작동마다계속바뀌지않게고정
                                                                               #coef=False라면 w를 빼줘야한다
# X, y, coef(기울기)  3개의 변수를 받는다

# X : [n_samples, n_features] 형상의 2차원 배열 ,독립 변수의 표본 데이터
# y : [n_samples] 형상의 1차원 배열 또는 [n_samples, n_targets] 형상의 2차원 배열, 종속 변수의 표본 데이터
# coef : [n_features] 형상의 1차원 배열 또는 [n_features, n_targets] 형상의 2차원 배열 (옵션). 선형 모형의 계수, 입력 인수 coef가 True 인 경우에만 출력됨

# 파라미터
#  n_samples : 정수 (옵션, 디폴트 100) 표본의 갯수
#  n_features : 정수 (옵션, 디폴트 100) 독립 변수(feature)의 수(차원)
#  n_targets : 정수 (옵션, 디폴트 1) 종속 변수(target)의 수(차원)
#  n_informative : 정수 (옵션, 디폴트 10) 독립 변수(feature) 중 실제로 종속 변수와 상관 관계가 있는 독립 변수의 수(차원)
#  effective_rank: 정수 또는 None (옵션, 디폴트 None) 독립 변수(feature) 중 서로 독립인 독립 변수의 수. 만약 None이면 모두 독립
#  tail_strength : 0부터 1사이의 실수 (옵션, 디폴트 0.5) effective_rank가 None이 아닌 경우 독립 변수간의 상관 관계 형태를 결정하는 변수
#  bias : 실수 (옵션, 디폴트 0.0) 절편
#  noise : 실수 (옵션, 디폴트 0.0) 출력 즉, 종속 변수에 더해지는 정규 분포의 표준 편차
#  coef : 불리언 (옵션, 디폴트 False) True 이면 선형 모형의 계수도 출력
#  random_state : 정수 (옵션, 디폴트 None) 난수 발생용 시작값

print(data.shape)
print()
print(y.shape)
print()
print(w)   # 기울기 w1
# y[1] - w *data[1]  =  96.3780614 (절편)
# y[0] - w *data[0] = 107.78034099

# y[1] - w * data[1]**2 = 92.90347204
# y[0] - w *data[0]**2 = 123.20618448
```
