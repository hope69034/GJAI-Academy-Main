```python
import pickle

class car:
    ui='''########### 자동차 정보 프로그램 ###########
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도 차 조회
6. 프로그램 이용 종료
############################################
입력:'''

    자동차_리스트=[]
    리스트최고속도=0
    리스트최저속도=999999999

    def __init__(self,이름,엔진가격,타이어가격,옵션유무,옵션이름,최고속도,구매가격):
        self.이름=이름
        self.엔진가격=엔진가격
        self.타이어가격=타이어가격
        self.옵션유무=옵션유무
        self.옵션이름=옵션이름
        self.최고속도=최고속도
        self.구매가격=구매가격
    
    def __repr__(self):
        return f'''이름:{self.이름}
엔진가격:{self.엔진가격}
타이어가격:{self.타이어가격}
옵션유무:{self.옵션유무}
옵션이름:{self.옵션이름}
최고속도:{self.최고속도}
구매가격:{self.구매가격}'''

    def 구매가조회(self):
        return f'''이름:{self.이름}
구매가격:{self.구매가격}'''

    @classmethod #인풋
    def f1(cls):
        이름=input('이름:')
        엔진가격=int(input('엔진가격:'))
        타이어가격=int(input('타이어가격:'))
        옵션유무=input('옵션유무:')
        if 옵션유무=='있음':
            옵션이름=input('옵션이름:')
        else:
            옵션이름=옵션유무
        최고속도=int(input('최고속도:'))
        구매가격=엔진가격+타이어가격
        자동차_정보=car(이름,엔진가격,타이어가격,옵션유무,옵션이름,최고속도,구매가격)
        cls.자동차_리스트.append(자동차_정보)
        cls.리스트최고속도=cls.리스트최고속도 if cls.리스트최고속도>최고속도 else 최고속도
        cls.리스트최저속도=cls.리스트최저속도 if cls.리스트최저속도<최고속도 else 최고속도
       
    @classmethod #목록
    def f2(cls):
        if len(cls.자동차_리스트)<=0:
            print('1대이상의차를등록하셔야함')
            return
        for i,자동차 in enumerate(cls.자동차_리스트): 
            print(f'''{i+1}번 차
{자동차}''')
    
    @classmethod #구매가
    def f3(cls):
        if len(cls.자동차_리스트)<=0:
            print('1대이상의차를등록하셔야함')
            return
        for i,자동차 in enumerate(cls.자동차_리스트): 
            print(f'''{i+1}번 차
{자동차.구매가조회()}''')  
            
    @classmethod #옵션차
    def f4(cls):
        if len(cls.자동차_리스트)<=0:
            print('1대이상의차를등록하셔야함')
            return
        for i,자동차 in enumerate(cls.자동차_리스트): 
            print(f'''{i+1}번 차
{자동차}''')

    @classmethod #속도차
    def f5(cls):
        if len(cls.자동차_리스트)<=1:
            print('2대이상의차를등록하셔야함')
            return
        print(f'속도차는 {cls.리스트최고속도-cls.리스트최저속도}입니다.')

    @classmethod
    def f6(cls):
        pickle.dump( cls.자동차_리스트,open('carinputlist' ,'wb')) #저장
        print('플그램종료')
        return True

    @classmethod
    def f7(cls):
        print('1~6입력해')

def main() :  #피쿨오픈 카자동차맇스트
    
    try:
        pickle.load(open('carinputlist' ,'rb')) #로드확인
        print('로드완료')
    except:
        print('로드실패 저장파일이없음')
    else:
        car.자동차_리스트=pickle.load(open('carinputlist' ,'rb')) #인풋리스트 로드
        print('인풋리스트를불러왔습니다')

    run=False

    while not run:
        n=int(input(car.ui))
        코드모델=[car.f1,car.f2,car.f3,car.f4,car.f5,car.f6,car.f7]
        if 1<=n<=6:
            run=코드모델[n-1]()
        else:
            코드모델[-1]()
#실행
if __name__=='__main__':
    main()


```
